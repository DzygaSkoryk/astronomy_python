### Reads FITS files with astropy

```python
from astropy.io import fits
import os

# find all fits files in a folder 
folder = '/home/dzyga/My/Python/Univer/PRG/NumPy_intro/'
filenames = [f for f in os.listdir(folder) if f.endswith('.fits')]
```

```python
fitsname = filenames[0]
hdulist = fits.open(folder + fitsname)

# read data from fits file
im = hdulist[0].data

plt.figure()
plt.imshow(im);
```

```python
# read all files 
DataCube = np.zeros([im.shape[0], im.shape[1], len(filenames)])
for i, name in enumerate(filenames):
    hdulist = fits.open(folder + name)
    onefits = hdulist[0].data
    DataCube[:,:,i] = onefits
```

```python
# find mean and std over 3d coordinate in the image stack
mean = np.mean(DataCube,axis=2)
std = np.std(DataCube,axis=2)
print(mean.shape)
print(std.shape)
```

```python
def running_stats(filenames):
    '''Calculates the running mean and stdev for a list of FITS files
    using Welford's method.'''
    n = 0
    for filename in filenames:
        hdulist = fits.open(filename)
        data = hdulist[0].data
        if n == 0:
            mean = np.zeros_like(data)
            s = np.zeros_like(data)

        n += 1
        delta = data - mean
        mean += delta/n
        s += delta*(data - mean)
        hdulist.close()

    s /= n - 1
    np.sqrt(s, s)

    if n < 2:
        return mean, None
    else:
        return mean, s
```
