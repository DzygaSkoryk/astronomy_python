```python
def my_function():
    print('Hello, folks!')
    return # this is optional


my_function()
```

```python
def power(x):
    return x**2

power(2)
```

```python
def digits(a, b=2):
    print('a', a, 'b', b)
    return a**2+b**2

digits(1)
digits(a=2)
digits(1, 3)
digits(a=2, b=3)
```

```python
# *args -- pass a variable number of non-keyword arguments
def add(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

add(2, 3)
add(2, 3, 3)
```

```python
# **kargs -- pass a keyworded variable-length arguments 

def total_fruits(**fruits):
    total = 0
    for key, value in fruits.items():
        print(key, value)
        total += value
    return total

total_fruits(banana=5, mango=7, apple=8)
```
