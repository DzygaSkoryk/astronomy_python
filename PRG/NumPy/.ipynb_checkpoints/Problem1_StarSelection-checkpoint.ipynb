```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# load image
import matplotlib.image as mpimg 
star = mpimg.imread('star.png') 

print(type(star))
  
# Output Images 
plt.figure()
plt.imshow(star); 
```

```python
plt.figure()
plt.imshow(star[:,:,3])
```

```python
star1 = star[:,:,0].copy() # take one channel

# see index of maximum element
ind = np.unravel_index(np.argmax(star1), star1.shape)
ind
```

```python
# create mask and fill background with nan value
star1[star1<0.5] = np.nan

plt.figure()
plt.imshow(star1)
```

```python
# mean value ignoring NaN 
np.nanmean(star1)
```
